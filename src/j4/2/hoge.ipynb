```python
import matplotlib.pyplot as plt

# Data
parallel = {"real": 2.04, "user": 0.01, "sys": 0.00}
wait = {"real": 18.23, "user": 0.01, "sys": 0.00}

# Categories
categories = ["real", "user", "sys"]
parallel_values = [parallel[cat] for cat in categories]
wait_values = [wait[cat] for cat in categories]

# Bar width
bar_width = 0.35

# Positions of the bars on the x-axis
r1 = range(len(categories))
r2 = [x + bar_width for x in r1]

# Create the figure and axes
fig, ax = plt.subplots()

# Plotting the bars
bar1 = ax.bar(r1, parallel_values, width=bar_width, label='parallel')
bar2 = ax.bar(r2, wait_values, width=bar_width, label='serial')

# Adding labels
ax.set_xlabel('Categories', fontweight='bold')
ax.set_ylabel('Time (s)', fontweight='bold')
ax.set_title('Parallel vs Serial Execution Times')
ax.set_xticks([r + bar_width / 2 for r in range(len(categories))])
ax.set_xticklabels(categories)
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
